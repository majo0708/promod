### mean() og std()

mean(hooyde) (importert fra pylab) gir oss gjennomsnittet av datasettet.

$$x = \frac{1}{n} \sum^{n}_{i=1} x_i$$

std(hooyde) (også importert fra pylab) gir oss standardavvik av datasettet, som er hvor mye avvik dataen har fra gjennomsnittet.

$$s = \sigma^2 = \sqrt{\frac{1}{n} \sum^{n}_{i=1} (x_i - x)^2}$$

```python
import pylab as p

data = p.loadtxt("46 statistikk.txt", skiprows=1, delimiter=",")

nr = data[:, 0]
hooyde = data[:, 1]

snitt = p.mean(hooyde)
avvik = p.std(hooyde)
p.plot(nr, hooyde)
p.xlabel("person")
p.ylabel("høyde")
p.show()

print("Snitt:", snitt, "Standardavvik:", avvik)

data2 = p.loadtxt("solflekker.txt", skiprows=1, delimiter="	")

nr2 = data2[:, 0]
hooyde2 = data2[:, 1]

p.plot(nr2, hooyde2)
p.xlabel("år")
p.ylabel("antall solflekker")
p.show()
```

### Regresjon
Regresjon lager en funksjon som passer bra til punktene man legger til.
Men regresjonsanalysen gir kanskje ikke en bra ekstrapolering (som vil si at funksjonen viser bra til andre x-verdier som ikke er punktene man la inn).

```python
T = [0, 20, 40, 60, 80, 100]
sol_NH3 = [88.5, 56.0, 34.0, 20.0, 11.0, 7.0]
sol_NaCl = [35.7, 35.9, 36.4, 37.1, 38.0, 39.2]

p.scatter(T, sol_NaCl)
#Regresjon
grad = 2
reg_NaCl = p.polyfit(T, sol_NaCl, grad)
#print(reg_NaCl)
x = p.linspace(0, 150, 10000)
y = reg_NaCl[0]*x**2 + reg_NaCl[1]*x + reg_NaCl[2]
#y_r2 = reg_NaCl[0]*T**2 + reg_NaCl[1]*T + reg_NaCl[2]

#r2 = r2_score(sol_NaCl, y_r2)
#print(r2)

p.plot(x, y, color='seagreen')

p.scatter(T, sol_NH3)
#Regresjon
grad = 2
reg_NH3 = p.polyfit(T, sol_NH3, grad)
#print(reg_NaCl)
x = p.linspace(0, 150, 10000)
a = reg_NH3[0]
b = reg_NH3[1]
c = reg_NH3[2]
y = a*x**2 + b*x + c

p.plot(x, y, color='magenta')
p.show()
```
