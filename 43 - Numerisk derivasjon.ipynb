# Numerisk derivasjon

### Funksjon for å derivere numerisk

```python
from pylab import *

def f(x):
    return x**2

def fder(f, x, dx=1E-8):
    der = (f(x + dx) - f(x))/dx
    return der

der = fder(f,1)

print(der)
```

### Analytisk derivasjon

En tangent er en linje som tangerer (skjærer gjennom et punkt i) en kurve. Stigningstallet $k$ til tangenten gjennom punktet $a$ kan skrives som funksjonen:

$$f'(a) = \lim_{\Delta a \to 0}\frac{f(a+\Delta a) - f(a)}{\Delta a}$$

Man kan bruke denne funksjonen fordi den viser når forskjellen på punktene nærmer seg null (når den blir en tangent), linja vil da tegne gjennom et punkt på buen og vi får en tangent.

Forskjellen mellom analytisk derivasjon og numerisk derivasjon er at analytisk derivasjon brukes av mennesker og man kan få eksakte verdier, men kan ikke brukes av datamaskiner. Derimot, kan maskiner bruke numerisk derivasjon, men vil få en liten feil siden man ikke kan tilnærme $\Delta x$ til 0.

# Dataen av diskrete data
### Derivering av strekning til fart, fra fart til akselerasjon

```python
from pylab import *
t = [5, 10, 15, 30, 50, 60]
s = [115, 190, 390, 600, 1000, 1200]
v = []
a = []
v.append(0)
a.append(0)

for i in range(len(s)-1):
    sder = (s[i+1] - s[i])/(t[i+1] - t[i])
    v.append(sder)
    vder = (v[i+1] - v[i])/(t[i+1] - t[i])
    a.append(vder)

plot(t, s)
show()
plot(t, v)
show()
plot(t, a)
show()
```

```python
import pylab as p

data = p.loadtxt("46 strekning ting.txt", skiprows=1, delimiter=",")

tid = data[:, 0]
hooyde = data[:, 2]
v = []
a = []
v.append(0)
a.append(0)

for i in range(len(hooyde)-1):
    sder = (hooyde[i+1] - hooyde[i])/(tid[i+1] - tid[i])
    v.append(sder)
    vder = (v[i+1] - v[i])/(tid[i+1] - tid[i])
    a.append(vder)

p.plot(tid, hooyde)
p.xlabel("tid (s)")
p.ylabel("høyde (m)")
p.show()
p.plot(tid, v)
p.xlabel("tid (s)")
p.ylabel("fart (m/s)")
p.show()
p.plot(tid, a)
p.xlabel("tid (s)")
p.ylabel("akselerasjon (m/s^2)")
p.show()
```
